```python
import pandas as pd
import pickle
```

## Read file and load to Pandas dataframe

```python
file='reading.sensor.pressuretransducer.pkl'
objects = []
with (open(file, "rb")) as openfile:
    while True:
        try:
            objects.append(pickle.load(openfile))
        except EOFError:
            break
```

```python
pd.DataFrame(objects)
```
